```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris.keys()
```

```python
df = pd.DataFrame(iris.data)
df.describe()
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df["label"] = iris.target
iris.target_names
```

```python
setosa =df[df["label"] == 0].drop("label", axis=1)
setosa.describe()
```

```python
versicolor = df[df["label"] == 1].drop("label", axis=1)
versicolor.describe()
```
